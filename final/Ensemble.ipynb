```python
from preprocessing import data
train_data,test_data,train_labels,test_labels = data()
```

```python
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
```

```python
X_train, X_val, y_train, y_val = train_test_split(train_data,train_labels, test_size=0.1, random_state=42)
```

```python
# Train k-NN
classifier = VotingClassifier(estimators=[
    ('lr', LogisticRegression()),
    ('svm', LinearSVC()),
    ('rf', RandomForestClassifier())
], voting='hard')
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_val)
```

```python
y_test_pred = classifier.predict(test_data)
```

```python
print("Classification Report:")
print(classification_report(test_labels, y_test_pred))
```

```python
a, b = 1, 2500
import random
random_integer = random.randint(a, b)

# Select the data point (a single row in the sparse matrix)
single_data_point = test_data[random_integer]

# Ensure the shape is correct (2D) if it's a single row
single_data_point = single_data_point.reshape(1, -1)

y_test_pred = classifier.predict(single_data_point)
```
