```python
from preprocessing import data
train_data,test_data,train_labels,test_labels = data()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
```

```python
X_train, X_val, y_train, y_val = train_test_split(train_data,train_labels, test_size=0.1, random_state=42)
```

```python
randomClassifier = RandomForestClassifier(n_estimators=450, random_state=42)
randomClassifier.fit(X_train, y_train)
```

```python
import numpy as np
import matplotlib.pyplot as plt

importances = randomClassifier.feature_importances_  # For RandomForest or GradientBoosting
indices = np.argsort(importances)

plt.bar(range(len(indices)), importances[indices],)
plt.title("Feature Importance")
plt.xlabel("Importance")
plt.ylabel("Feature Index")
plt.show()
```

```python
y_pred = randomClassifier.predict(X_val)
```

```python
y_test_pred = randomClassifier.predict(test_data)
```

```python
print("Classification Report:")
print(classification_report(test_labels, y_test_pred))
```

```python
a, b = 1, 2500
import random
random_integer = random.randint(a, b)

# Select the data point (a single row in the sparse matrix)
single_data_point = test_data[random_integer]

# Ensure the shape is correct (2D) if it's a single row
single_data_point = single_data_point.reshape(1, -1)

y_test_pred = randomClassifier.predict(single_data_point)
```
